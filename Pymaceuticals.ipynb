```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
# mouse_drug_data = "mouse_drug_data.csv"
# clinical_trial_data = "clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv("mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("clinicaltrial_data.csv")

# Combine the data into a single dataset
# DataFrame.merge(self, right, how='outer', on=None, left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)

combined_data = pd.merge(clinical_trial_data, mouse_drug_data, how='left', on="Mouse ID", left_on=None, right_on=None, left_index=False, right_index=False, sort=False, suffixes=('_x', '_y'), copy=True, indicator=False, validate=None)

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_tumor_vol = combined_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].mean()
type(avg_tumor_vol)

# Convert to DataFrame
avg_tumor_vol_df = pd.DataFrame(avg_tumor_vol)

# Preview DataFrame
avg_tumor_vol_df = avg_tumor_vol_df.reset_index()
avg_tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_tumor_error = combined_data.groupby(['Drug', 'Timepoint'])['Tumor Volume (mm3)'].sem()
type(standard_tumor_error)

# Convert to DataFrame
standard_tumor_error_df = pd.DataFrame(standard_tumor_error)

# Preview DataFrame
standard_tumor_error_df = standard_tumor_error.reset_index()
standard_tumor_error_df.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_tumor_vol = avg_tumor_vol.reset_index()
avg_tumor_vol_pivot = avg_tumor_vol.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

standard_tumor_error = standard_tumor_error.reset_index()
standard_tumor_error_pivot = standard_tumor_error.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
avg_tumor_vol_pivot.head()
standard_tumor_error_pivot.head()
```

```python
# Generate the Plot (with Error Bars) 
tumor_vol_df = avg_tumor_vol_pivot[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
tumor_error_df = standard_tumor_error_pivot[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]

tumor_vol_df.plot(kind='line', use_index=True, grid=True, yerr=tumor_error_df, marker='.', linewidth=1)

plt.title('Tumor Response to Treatment')
plt.xlabel('Time (days)')
plt.ylabel('Tumor Volume (mm3)')

plt.legend(loc="best", fontsize="medium", fancybox=True)

# Save the Figure
plt.savefig("Response.png")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avg_met_site = combined_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].mean()
type(avg_met_site)

# Convert to DataFrame
avg_met_site_df = pd.DataFrame(avg_met_site)

# Preview DataFrame
avg_met_site_df = avg_met_site_df.reset_index()
avg_met_site_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
standard_metsite_error = combined_data.groupby(['Drug', 'Timepoint'])['Metastatic Sites'].sem()
type(standard_metsite_error)

# Convert to DataFrame
standard_metsite_error_df = pd.DataFrame(standard_metsite_error)

# Preview DataFrame
standard_metsite_error_df = standard_metsite_error.reset_index()
standard_metsite_error_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
avg_met_site = avg_met_site.reset_index()
avg_met_site_pivot = avg_met_site.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

standard_metsite_error = standard_metsite_error.reset_index()
met_site_error_pivot = standard_metsite_error.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
avg_met_site_pivot.head()
met_site_error_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
# scatter plot that shows how the number of metastatic (cancer spreading) sites changes over time for each treatment.

met_site_df = avg_met_site_pivot[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
met_error_df = met_site_error_pivot[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]

met_site_df.plot(kind='line', use_index=True, grid=True, yerr=met_error_df, marker='.', linewidth=1)

plt.title('Metastatic Spread During Treatment')
plt.xlabel('Treatment Duration (days)')
plt.ylabel('Met Sites')

plt.legend(loc="best", fontsize="medium", fancybox=True)

# Save the Figure
plt.savefig("Spread.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count = combined_data.groupby(['Drug', 'Timepoint'])['Mouse ID'].count()
type(mice_count)

# Convert to DataFrame
mice_count_df = pd.DataFrame(mice_count)

# Preview DataFrame
mice_count_df = mice_count_df.reset_index()
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count = mice_count.reset_index()
mice_count_pivot = mice_count.pivot(index="Timepoint", columns="Drug")["Mouse ID"]

# Preview the Data Frame
mice_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
# scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)

mice_count_df = mice_count_pivot[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
 
mice_count_df.plot(kind='line', use_index=True, grid=True, marker='*', linewidth=1)

plt.title('Survival During Treatment')
plt.xlabel('Survival Rate (%)')
plt.ylabel('Time (Days)')

plt.legend(loc="best", fontsize="medium", fancybox=True)

# Save the Figure
plt.savefig("Survival.png")

# Show the Figure
plt.show()
```
